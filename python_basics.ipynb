### Slicing Methods

```python
# slicing a list, 
my_list = [7, 12, 3, 9, 5, "apple", "banana", "cherry"]

# slicing with first start element want to be included, end element of the first element you want to exclude
print(my_list[1:4])

# slicing with negative values
# last value will be mylist[-1] there is no zero index in negative indices

print(my_list[-3:-1])
print(my_list[-1:-3:-1]) # equivalent, make sure to indicate the step size

# slicing with steps
print(my_list[0:8:2])

# use slicing to modify lists
my_list_1 = my_list.copy()
my_list_1[1:5] = ['orange','strawberry','grape','pear']
print(my_list_1)
```

```python
# copy a python list
my_list_copy = my_list[:]
print(my_list_copy)

# reverse a list
reversed_list = my_list[::-1]
print(reversed_list)

# get the first or last n elements
print(my_list[:3]) # first three
print(my_list[-3:]) # last three
print(my_list[:-3]) # remove last three from the original list
```

```python
# remove duplicate values
lst = [1,2,2,3,4,4,5,6,6,7]
print(list(set(lst))) # set will generate {1,2,3,4,5,6,7}
```

```python
# example: find palindromic substrings
def palindrome(x):
    subs = []
    output = []
    for i in range(len(x)):
        for j in range(i+1,len(x)+1):
            subs.append(x[i:j])
    for sub in subs:
        if sub == sub[::-1]:
            output.append(sub)
    return output
```

```python
input_str = "abcbamadam"
palindrome(input_str)
```

### List Operations

```python
# append an element to the list, add an element at the end of the list
my_list = [1,2,3,4,5]
my_list.append(6)
print(my_list)

# insert an element at a specific index
my_list_copy = my_list[:]
my_list_copy.insert(2,8)
print(my_list_copy)

# remove an element, check if the element is in the list
my_list.remove(5)
print(my_list)

# reverse the list
my_list.reverse()
print(my_list)

# sort the list
my_list.sort(reverse=False) #reverse means in ascending or decending order
print(my_list)
```

## Dictionary

### Accessing Dictionary Iterms

```python
# creating a dictionary
my_dict = {'key1':'value1','key2':'value2','key3':'value3'}
print(my_dict)

# creating adictionary from adding items
my_dict = {}
my_dict['key1'] = 'value1'
my_dict['key2'] = 'value2'
my_dict['key3'] = 'value3'
print(my_dict)

# accessing the value at dictionary
print(my_dict['key1'])

# accessing the value by get()
print(my_dict.get('key1'))

# if the value does not exist
print(my_dict.get('key12'))

# provide the default value
print(my_dict.get('key12','value*'))
```

## Mutable vs Immutable
In python, objects are created can be either mutable or immutable.

```python
# for immutable objects
a = 5 # a points to 5 which is immutable
b = a # b points to 5
a = 6 # a points to a new object which is 6
print(a)
print(b) # b is still pointing at 5
```

```python
# for mutable objects like lists and dictionary
a = [1,2,3]
b = a # b points to the same object as a
a.append(4) # modify the a
print(a)
print(b)
```
