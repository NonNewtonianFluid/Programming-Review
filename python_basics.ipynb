### Slicing Methods

````python
# slicing a list, 
my_list = [7, 12, 3, 9, 5, "apple", "banana", "cherry"]

# slicing with first start element want to be included, end element of the first element you want to exclude
print(my_list[1:4])

# slicing with negative values
# last value will be mylist[-1] there is no zero index in negative indices

print(my_list[-3:-1])
print(my_list[-1:-3:-1]) # equivalent, make sure to indicate the step size

# slicing with steps
print(my_list[0:8:2])

# use slicing to modify lists
my_list_1 = my_list.copy()
my_list_1[1:5] = ['orange','strawberry','grape','pear']
print(my_list_1)
````

````python
# example: find palindromic substrings
def palindrome(x):
    subs = []
    output = []
    for i in range(len(x)):
        for j in range(i+1,len(x)+1):
            subs.append(x[i:j])
    for sub in subs:
        if sub == sub[::-1]:
            output.append(sub)
    return output
````

## Mutable vs Immutable
In python, objects are created can be either mutable or immutable.

````python
# for immutable objects
a = 5 # a points to 5 which is immutable
b = a # b points to 5
a = 6 # a points to a new object which is 6
print(a)
print(b) # b is still pointing at 5
````

````python
# for mutable objects like lists and dictionary
a = [1,2,3]
b = a # b points to the same object as a
a.append(4) # modify the a
print(a)
print(b)
````

### Variable-Length Parameters
Python allows us to define variable-length parameters, which are represented by an asterisk. This allows us to pass arbitrary number of input values to a function. 

````python
"""
def function_name(*args):
    # code block to be executed when the function is called
    return value
"""
def sum(*numbers):
    results = 0
    for num in numbers:
        results += num
    return results

print(sum(1,2,3,4,5,6))
print(sum(1,2,3,4))
````

````python
def moveZeroes(nums):
    """
    :type nums: List[int]
    :rtype: None Do not return anything, modify nums in-place instead.
    """
    result = []
    index = []
    for i in range(len(nums)):
        if nums[i] != 0:
            result.append(nums[i])
        else:
            index.append(0)
    return result + index
moveZeroes([0,1,0,3,12])
````
